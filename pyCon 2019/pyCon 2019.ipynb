## 2. Which talks provoke the most online discussion?

```python
ted.sort_values('comments').tail()
```

```python
ted['comments_per_view'] = ted.comments / ted.views
```

```python
ted.sort_values('comments_per_view').tail()
```

```python
ted['comments_per_1000view'] = ted.comments_per_view * 1000
```

```python
ted.sort_values('comments_per_view').tail()
```

```python
ted['view_per_comments'] = ted.views / ted.comments
```

```python
ted.sort_values('view_per_comments').head()
```

```python
ted.drop(columns = ['comments_per_view', 'comments_per_1000view']).head()
```

## 3. Visualize the distrubution of comments

```python
ted[ted.comments<1000].shape
```

```python
ted[ted.comments<1000].comments.plot(kind = 'hist')
```

```python
ted.query('comments < 1000').comments.plot(kind = 'hist') # Using query
```

```python
ted.loc[ted.comments < 1000, 'comments'].plot(kind = 'hist', bins = 20)
```
